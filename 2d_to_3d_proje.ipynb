```python
# Create a fake gpytoolbox module so imports don't crash
stub_path = "/content/BE_PROJE_2D_TO_3D/sv3d_repo/sf3d/models/gpytoolbox.py"

with open(stub_path, "w") as f:
    f.write("""
# Stub gpytoolbox for SV3D (Colab-safe)
class Dummy:
    def __getattr__(self, name):
        return lambda *args, **kwargs: None
""")

print("✅ gpytoolbox stub created")
```

```python
mesh_file = "/content/BE_PROJE_2D_TO_3D/sv3d_repo/sf3d/models/mesh.py"

with open(mesh_file, "r") as f:
    lines = f.readlines()

new_lines = []
for line in lines:
    if line.strip() == "import gpytoolbox":
        new_lines.append(
            "try:\n"
            "    import gpytoolbox\n"
            "except ImportError:\n"
            "    from . import gpytoolbox\n"
        )
    else:
        new_lines.append(line)

with open(mesh_file, "w") as f:
    f.writelines(new_lines)

print("✅ mesh.py updated")
```
